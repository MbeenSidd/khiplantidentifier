```python
import os
from bs4 import BeautifulSoup
import urllib.request
from PIL import Image

search_queries = ['snake plant', 'money plant', 'aloe vera plant', 'white bougainvillea plant', 'maggie plant', 'pandanus urophyllus plant', 'swiss cheese plant','peregrina plant' ,'song of india plant', 'dumb cane plant', 'damas plant']
#search_queries =['peregrina plant']
for query in search_queries:
    query = query.split()
    query = '+'.join(query)
    url = f"https://www.bing.com/images/search?q={query}"

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
    req = urllib.request.Request(url, headers=headers)
    html = urllib.request.urlopen(req).read()

    try:
        os.mkdir(query)
    except:
        pass

    soup = BeautifulSoup(html, 'html.parser')
    urls = []
    for img in soup.findAll('img'):
        try:
            url = img['data-src']
            urls.append(url)
        except:
            try:
                url = img['src']
                urls.append(url)
            except:
                pass

    for i, url in enumerate(urls):
        if i > 10000:
            break
        filename = os.path.join(query, f"{query}{i}.jpg")
        try:
            urllib.request.urlretrieve(url, filename)
            print(f"Downloading {filename}")
            
            # Check minimum image size
            img = Image.open(filename)
            width, height = img.size
            if width <= 512 or height <= 512:
                os.remove(filename)
                print(f"Deleted {filename} due to smaller dimensions.")
        except:
            print(f"Failed to download {filename}")

```
